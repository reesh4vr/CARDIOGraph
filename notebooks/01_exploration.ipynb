Use this notebook to inspect the raw biomedical tables after they are cleaned by `loader.data_loader`. Make sure the five input files are placed inside `data/raw/` before running any cells.



```python
from pathlib import Path
import sys

# Resolve project root (one level up from notebooks/)
project_root = Path.cwd().parents[0]  # from notebooks/ â†’ project root
sys.path.append(str(project_root / "src"))

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from loader.data_loader import load_all_data

sns.set_theme(style="whitegrid")
```

```python
data = load_all_data()
print("Datasets loaded:")
for name, df in data.items():
    print(f"- {name}: {df.shape[0]:,} rows x {df.shape[1]} cols")
```

```python
for name, df in data.items():
    display(pd.DataFrame({"dataset": [name], "rows": [df.shape[0]], "cols": [df.shape[1]]}))
    display(df.head())
    display(df.describe(include="all"))
    print("-" * 80)

```
